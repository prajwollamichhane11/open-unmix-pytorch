```python
import matplotlib.pyplot as plt
import json
import os
import numpy as np
```

```python
path = 'open-unmix-512/podcasts.json'
```

```python
with open(path) as handle:
    data = json.load(handle)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax.plot(data['train_loss_history'], label='train', color='g')
ax.plot(data['valid_loss_history'], label='valid', color='m')
ax.scatter(data['best_epoch']-1, data['best_loss'],
           s=500, marker='v', color='c')
plt.legend(prop={'size': 26})
plt.ylim([0.0, 1.0])
ax.set_ylabel('mse loss', size=26)
ax.set_xlabel('epochs', size=26)
ax.tick_params(axis='both', which='major', labelsize=20)
plt.show()
```
